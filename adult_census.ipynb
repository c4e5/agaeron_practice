```python
"""
Bins for education
0 - Preschool --> HS-grad
1 - Some-college + Assoc-acdm + Assoc-voc
2 - Bachelors
3 - Masters
4 - Doctorate + Prof-school
"""
df["education-binned"] = df["education"].map({" Doctorate":4, " Prof-school":4, " Bachelors":2, " Masters":3, " Some-college":1,\
                                                    " Assoc-acdm":1, " Assoc-voc":1, " Preschool":0, \
                                                    " 1st-4th":0, " 5th-6th":0, " 7th-8th":0, " 9th":0, \
                                                    " 10th":0, " 11th":0, " 12th":0, " HS-grad":0}).values

"""
0 - Separated + Never-married
1 - Divorced + Widowed + Married-spouse-absent
2 - Married-AF-spouse + Married-civ-spouse

"""
df["marital-status-binned"] = df["marital-status"].map({" Separated":0, " Never-married":0, \
                                                      " Divorced":1, " Widowed":1, " Married-spouse-absent":1,\
                                                      " Married-AF-spouse":2, " Married-civ-spouse":2}).values

"""
0 - without-pay + Never-worked
1 - ? (this will be 'undefined' which could include nonspecific or rare jobs) 
2 - Private            
3 - Local-gov + Self-emp-not-inc + State-gov 
4 - Federal-gov
5 - Self-emp-inc
"""

df["workclass-binned"] = df["workclass"].map({" Without-pay":0, " Never-worked":0, " ?":1, " Private":2,\
                                             " Local-gov":3, " Self-emp-not-inc":3, " State-gov":3," Federal-gov":4,\
                                             " Self-emp-inc":5}).values
# race binning
"""
0 - Amer-Indian-Eskimo + Black + Other
1 - White + Asian-Pac-Islander
"""
df["race-binned"] = df["race"].map({" Amer-Indian-Eskimo":0, " Black":0, " Other":0,\
                                     " White":1, " Asian-Pac-Islander":1}).values

# sex binning
"""
0 - Female
1 - Male
"""
df["sex-binned"] = df["sex"].map({" Female":0, " Male":1}).values

"""
0 - countries other than the US
1 - US
"""
df["native-country-binned"] = df["native-country"].values == " United-States"
df["native-country-binned"] = df["native-country-binned"].astype(int)
#df["native-country"].value_counts()

# relationship binning
"""
0 - own-child + other-relative + unmarried
1 - not-in-family
2 - husband + wife
"""
df["relationship-binned"] = df["relationship"].map({" Own-child":0, " Other-relative":0, " Unmarried":0, \
                                                    " Not-in-family":1, " Husband":2, " Wife":2}).values
#Occupation binning
"""
0 - Priv-house-serv + Other-service + Handlers-cleaners
1 - ? + Armed-Forces + Farming-fishing + Machine-op-inspct + Adm-clerical
2 - Transport-moving + Craft-repair 
3 - Sales 
4 - Tech-support + Protective-serv
5 - Exec-managerial + Prof-specialty
"""

df["occupation-binning"] = df["occupation"].map({" Priv-house-serv": 0, ' Other-service': 0, ' Handlers-cleaners': 0, \
                                                ' ?': 1, ' Armed-Forces': 1, ' Farming-fishing': 1, ' Machine-op-inspct': 1, \
                                                ' Adm-clerical': 1, ' Transport-moving': 2, ' Craft-repair': 2, ' Sales': 3, \
                                                ' Tech-support':4, ' Protective-serv':4, ' Exec-managerial':5, \
                                                ' Prof-specialty':5})
```

```python
df_binned = shuffle(df[["age", "workclass-binned", "fnlwgt", "education-binned",\
                'native-country-binned', 'education-binned','marital-status-binned', 'workclass-binned',\
                'race-binned', 'sex-binned', 'relationship-binned', 'occupation-binning', 'hours-per-week', \
                'capital-gain', 'capital-loss', "MONEY"]])
```
